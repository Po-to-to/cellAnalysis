```python
class MyDataFrame(pd.DataFrame):
    
    def __init__(self, data, user):
        super().__init__(data)
        
        self.user = user
        self.hello()
    
    def imshow(self):
        plt.imshow(self.values)
        print(self.user)
        self.hello()
        
    def hello(self):
        print("Hello")
```
